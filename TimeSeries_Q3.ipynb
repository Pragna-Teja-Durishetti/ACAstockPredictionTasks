-> The lag operator shifts a time series back by a specified number of time steps
->It helps identify relationships between past and present values which is very important for time series modeling.

```python
import numpy as np
import matplotlib.pyplot as plt


sales_data = [20, 24, 22, 26, 25, 28, 30, 27, 29, 31]
N = len(sales_data)

sum_of_sales = 0
for sale in sales_data:
    sum_of_sales = sum_of_sales + sale

mean_sales = sum_of_sales / N


def calculate_autocorr_manual(data_list, lag_value):

    num_points = len(data_list)
    data_mean = sum(data_list) / num_points


    numerator_sum = 0

    for i in range(num_points - lag_value):
        term1 = data_list[i] - data_mean
        term2 = data_list[i + lag_value] - data_mean
        product = term1 * term2
        numerator_sum = numerator_sum + product


    denominator_sum = 0
    for i in range(num_points):
        term = data_list[i] - data_mean
        squared_term = term * term
        denominator_sum = denominator_sum + squared_term

    autocorr_coefficient = numerator_sum / denominator_sum
    return autocorr_coefficient

autocorr_lag1 = calculate_autocorr_manual(sales_data, 1)
print(f"  Autocorrelation at Lag 1: {autocorr_lag1}\n")

autocorr_lag2 = calculate_autocorr_manual(sales_data, 2)
print(f"  Autocorrelation at Lag 2: {autocorr_lag2}\n")


lags_to_plot = [1, 2]
autocorr_values_to_plot = [autocorr_lag1, autocorr_lag2]



plt.figure(figsize=(8, 5))
plt.bar(lags_to_plot, autocorr_values_to_plot, color=['blue', 'red'], width=0.4)


plt.axhline(0, color='grey', linestyle='--', linewidth=0.8)


plt.title('Autocorrelation for Lags 1 and 2 of Sales Data')
plt.xlabel('Lag (Number of periods back)')
plt.ylabel('Autocorrelation Coefficient')

plt.xticks(lags_to_plot, ['Lag 1', 'Lag 2'])


plt.ylim(-1, 1)


plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

->Autocorrelation for lag1 is 0.42437. This is a moderate positive value. It suggests that there's some tendency for sales to follow the previous day's pattern.

->Autocorrelation for Lag 2: 0.3810. This is a moderate positive value. It suggests that there's some tendency for sales to follow the previous day's pattern.

A stationary time series has
->Constant mean

->Constant variance

->No trend or seasonality

```python
#Making a time series stationary:

import pandas as pd


Arr = [20, 24, 22, 26, 25, 28, 30, 27, 29, 31]

# Convert to Series
series = pd.Series(Arr)

# Apply first-order differencing
StArr = series.diff()

# Display the differenced (stationary) series
print(StArr)
```
